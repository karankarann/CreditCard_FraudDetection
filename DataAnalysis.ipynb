### Import Libaries

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

### Getting the Data

```python
df = pd.read_csv("Data/creditcard.csv.zip")
```

```python
df.head()
```

### Analysis

```python
df.info()
```

```python
df.describe()
```

```python
## Making sure there are only 2 Classes in the prediction class "0" and "1"
df["Class"].unique() 
```

```python
## Checking Null values
df.isnull().sum() != 0
```

```python
## Distribution of the Prediction Classes
print("Prediction Class Total {}".format( len(df) ) )
for i in range(0,2):
    print("Prediction Class {0}: {1} -> {2:2f}%".format(i, len(df[df["Class"] == i]) , (len(df[df["Class"] == i])/len(df))*100) )
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(), cmap="Blues", linecolor="Black", linewidths="1")
```
